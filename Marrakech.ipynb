```python
from requests_html import HTMLSession
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

s = HTMLSession()
url = 'https://www.tripadvisor.com/Search?q=Marrakech&searchSessionId=001e7466d6fd3aff.ssid&sid=EA94DCA1F92343F1A7BB033A779B6E3D1688575427245&blockRedirect=true&ssrc=h&isSingleSearch=true&geo=1'
```

```python
def gerdata(url):
    driver = webdriver.Chrome()
    wait = WebDriverWait(driver, 30)
    
    driver.get(url)
    
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'result-title')))
    
    html = driver.page_source 
    driver.quit() 
    
    soup = BeautifulSoup(html)
    return soup
```

```python
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 30)
driver.get(url)
url_list = []
while True:
    next_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "ui_button.nav.next.primary")))
    current_url = driver.current_url
    url_list.append(current_url)
    next_button.click()
    if driver.current_url == current_url:
        break
    if len(url_list)==33:
        break
for url in url_list:
    print(url)
driver.quit()
```

```python
def starts_with_ui_bubble_rating(class_name):
    return class_name and class_name.startswith("ui_bubble_rating bubble_")
```

```python
def getdata(html):
    hotelsdiv=html.find_all("div", class_='ui_columns is-mobile result-content-columns')
    for i in range(0,len(hotelsdiv)):
        hotel=[]
        div_hotel_element = hotelsdiv[i].find("div", class_="result-title")
        span_element = div_hotel_element.find("span").text
        hotel.append(span_element)
        if hotelsdiv[i].find_all("span", class_=starts_with_ui_bubble_rating)!=[]:
            div_bubble_rating_element = hotelsdiv[i].find("span", class_=starts_with_ui_bubble_rating)
            rating = div_bubble_rating_element.get("alt").split()
            hotel.append(rating[0])
        else:
            rating = hotelsdiv[i].find("a", class_="write-review").text
            hotel.append(rating)
        location_element = hotelsdiv[i].find("div", class_="address-text").text
        hotel.append(location_element)
        hotels.append(hotel)
```

```python
len(url_list)
url_list1=url_list[:15]
url_list2=url_list[15:25]
url_list3=url_list[25:]
```

```python
hotels=[]
for url in url_list1:
    print(url_list.index(url))
    html = gerdata(url)
    getdata(html)
    
hotels
```
